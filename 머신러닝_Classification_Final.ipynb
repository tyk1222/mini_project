```python
import pandas as pd
```

```python
# 데이터 불러오기
train_data = pd.read_csv('/content/drive/MyDrive/머신러닝/playground-series-s3e2/train.csv')
test_data = pd.read_csv('/content/drive/MyDrive/머신러닝/playground-series-s3e2/test.csv')
submission = pd.read_csv('/content/drive/MyDrive/머신러닝/playground-series-s3e2/sample_submission.csv')
```

```python
# 훈련 데이터 확인
train_data.head()
```

```python
train_data.info()
```

```python
train_data.describe()
```

```python
# 학습에 불필요한 id컬럼 제거
X = train_data.drop(['id', 'Residence_type'], axis=1)
y = train_data['stroke']

X_ = test_data.drop(['id', 'Residence_type'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 데이터분할
X_train, X_test, y_train, y_test = train_test_split( X.iloc[:, :-1],
                                X['stroke'], test_size=0.15, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# 데이터 X준비
X = pd.concat([X_train, y_train], axis = 1)
X.shape
```

```python
# from pycaret.regression 회귀용
from pycaret.classification import *
# 분류용 모든 모듈을 가져온다
```

```python
from category_encoders.target_encoder import TargetEncoder
clf = setup( data   = X,                # 훈련데이터 => DataFrame로 공급
             target = 'stroke',         # 정답 데이터의 컬럼명
             verbose= True,             # 로그출력
             train_size = 0.9,          # 훈련:검증 비율시 90%는 훈련에 사용 => cv
             data_split_shuffle=True,   # 데이터를 나눌때 섞음
             session_id = 100,          # 난수 시드 100
             normalize  = False,        # 정규화 처리 안함
             )
```

```python
models()
```

```python
# 대회 평가 항목에 Submissions are evaluated on area under the ROC curve between the predicted
#                   probability and the observed target 라는 한 줄짜리 문구가 있는데
# "번역시 제출물은 예측된 확률과 실제 대상 간의 ROC 곡선 아래 영역(Area Under the ROC Curve)을 기반으로 평가됩니다"
# 가 있으므로 'AUC' 설정
top5_model = compare_models(fold = 5, round = 3, sort = 'AUC', n_select = 5)
```

```python
# 상위 5개 모델 확인
top5_model
```

```python
blended = blend_models(
            estimator_list=top5_model,  # 상위 5개 모델을 놓고
            fold=5,                     # 폴드 설정
            #method='hard',             # 보팅방식
            #optimize='AUC'             # 랭킹산정기준, 판단지표
            )
```

```python
# 파이프라인 구성
model = finalize_model(blended)
model
```

```python
y_pred = predict_model(model, data=X_)
y_pred
```

```python
# 'prediction_label' 의 0과 1의갯수 확인
y_pred['prediction_label'].value_counts()
```

```python
# ex) 양성:1을 기준으로 놓고 양성일 경우 예측포인트는 높게, 0일경우 100% 확률에서 prediction_score 값을 빼는 방법으로
# 한쪽을 기준으로 잡고 확률을 반전시킴
submission['stroke']=list(map(lambda x, y : 1-y if x == 0 else y, y_pred['prediction_label'], y_pred['prediction_score']))
submission.head()
```
